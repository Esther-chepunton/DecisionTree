```python
#load the dataset
import pandas as pd
animal= pd.read_csv('/content/animals.csv')
```

```python
animal
```

```python
class_mapping = {
    1: 'Mammals',
    2: 'Birds',
    3: 'Reptiles',
    4: 'Fish',
    5: 'Amphibians',
    6: 'Insects',
    7: 'Arachnids'
}
```

```python
animal['Class'] = animal['Class'].map(class_mapping)
```

```python
animal
```

```python
animal['Class']=animal['Class'].replace(['Reptiles','Fish','Amphibians','Birds','Insects','Arachnids'],'Non-Mammals')
```

```python
animal
```

```python
#pd.crosstab([animal['Hair'],animal['Milk']], animal['Class'])
```

```python
#decision trees
from sklearn import tree
```

```python
y=animal['Class']
```

```python
y
```

```python
x=animal.drop(['Sr','Class'],axis=1)
```

```python
x
```

```python
clf=tree.DecisionTreeClassifier(criterion='entropy',max_depth=3)
```

```python
clf
```

```python
clf=clf.fit(x,y)
```

```python
clf
```

```python
import pydotplus
```

```python
from IPython.display import Image
```

```python
dot_animal=tree.export_graphviz(clf,feature_names=x.columns,class_names=['Mammals','Non-Mammals'], filled=True,out_file=None)
```

```python
dot_animal
```

```python
#convert to graph
graph=pydotplus.graph_from_dot_data(dot_animal)
```

```python
graph
```

```python
testanimal=[['91',0,0,0,0,1,1,'non-mammals'],
         ['92',1,0,0,0,1,1,'mammals'],
         ['93',1,0,0,1,1,0,'non-mammals'],
         ['94',1,1,1,0,0,0,'mammals'],
         ]
```

```python
# Convert testanimal to a DataFrame with only the relevant columns
testanimal_df = pd.DataFrame(testanimal, columns=['Sr', 'Hair', 'Feathers', 'Eggs', 'Milk', 'Airborne', 'Aquatic', 'Class'])
```

```python
# Merge the two DataFrames on the 'Sr' column using pd.concat
merged_df = pd.concat([animal, testanimal_df], ignore_index=True)
```

```python
# Rename the 'Class' column in testanimal_df to 'Class_new' within the merged DataFrame
merged_df = merged_df.rename(columns={'Class': 'Class_new'})
```

```python
# Update the 'Class' column in the original DataFrame with the new 'Class' values from testanimal
animal['Class'] = merged_df['Class_new'].combine_first(animal['Class'])
```
